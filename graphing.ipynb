```python
# Import numpy and pyplot 
import matplotlib.pyplot as plt
import numpy as np
from __future__ import print_function
```

```python
#static graph
%matplotlib inline
#%matplotlib notebook #interactive graph
```

```python
#preferred spatial frequency graph

n1 = 20
pref_sf_sg = .2+.1*np.random.randn(n1) #arbitrary random numbers within axis range
pref_sf_natimages = .1+.1*np.random.randn(n1) #arbitrary random numbers within axis range
size1 = 100

#make figure,axes handles
fig,ax=plt.subplots()

#plot scatter plots
ax.scatter(pref_sf_sg,pref_sf_natimages,s=size1,color='red',alpha=0.25)
#ax.scatter(x2,y2,s=size2,color='blue',alpha=0.25)

#set graph title
fig.suptitle('Preferred spatial frequency',fontsize=18);

#set x axis label & limits
ax.set_xlabel("Static Gratings", fontsize=14)
ax.set_xlim(0,.32)
ax.set_xticks([.02, .04, .08, .16, .32]);
#ax.set_xscale('log',basex=2) #make x axis log base 2 scale

#set y axis label & limits
ax.set_ylabel("Natural Images", fontsize=14)
ax.set_ylim(0,.32)
ax.set_yticks([.02, .04, .08, .16, .32]);
#ax.set_yscale('log',basex=2) #make y axis log base 2 scale
```

```python
#preferred orientation graph

n2 = 20
pref_ori_sg = 50+40*np.random.randn(n2) #arbitrary random numbers within axis range
pref_ori_natimages = 100+30*np.random.randn(n2) #arbitrary random numbers within axis range
size1 = 100

#make figure,axes handles
fig,ax=plt.subplots()

#plot scatter plots
ax.scatter(pref_ori_sg,pref_ori_natimages,s=size1,color='green',alpha=0.25)

#set graph title
fig.suptitle('Preferred orientation',fontsize=18);

#set x axis label & limits
ax.set_xlabel("Static Gratings", fontsize=14)
ax.set_xlim(0,150)
ax.set_xticks([30, 60, 90, 120, 150]);

#set y axis label & limits
ax.set_ylabel("Natural Images", fontsize=14)
ax.set_ylim(0,150)
ax.set_yticks([30, 60, 90, 120, 150]);
```

```python
# Create figure and axes objects. Make them share the x-axis
fig, axes = plt.subplots(nrows=4, ncols=1, sharex = True, figsize=(10,10))

# Plot on each subplot by indexing into 'ax'
axes[0].scatter(pref_ori_sg,pref_ori_natimages, label='V1')
axes[1].scatter(pref_ori_sg,pref_ori_natimages, label='PM', color='red')
axes[2].scatter(pref_ori_sg,pref_ori_natimages, label='AL', color='green')
axes[3].scatter(pref_ori_sg,pref_ori_natimages, label='L', color='yellow')

axes[0].legend(loc='upper right')
axes[1].legend(loc='upper right')
axes[2].legend(loc='upper right')
axes[3].legend(loc='upper right')

#set graph title
fig.suptitle('Preferred Orientation',fontsize=22, y=.95);

#set x axis label & limits
axes[3].set_xlabel("Static Gratings", fontsize=18)
axes[0].set_xlim(0,150)
axes[0].set_xticks([30, 60, 90, 120, 150])

#set y axis label & limits
axes[0].set_ylabel("Natural Images", fontsize=18)
axes[0].yaxis.set_label_coords(-.06, -1.2) 

axes[0].set_ylim(0,150)
axes[1].set_ylim(0,150)

axes[0].set_yticks([30, 60, 90, 120, 150]);
axes[1].set_yticks([30, 60, 90, 120, 150]);
axes[2].set_yticks([30, 60, 90, 120, 150]);
axes[3].set_yticks([30, 60, 90, 120, 150]);
```
